```python
#라이브러리 실행
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
browser = webdriver.Chrome('C:/r-selenium/chromedriver.exe')
```

```python
url = 'https://www.melon.com/chart/index.htm'
browser.get(url)
browser.implicitly_wait(3)
```

```python
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
song_list = soup.select('tbody > tr')
```

```python
song = song_list[0]
song
```

```python
#곡명, 가수, 앨범명, 좋아요수 가져오기
title = song.select('div.ellipsis.rank01')[0].text.replace('\n', '')
singer = song.select('div.ellipsis.rank02 > a')[0].text
album = song.select('div.ellipsis.rank03 > a')[0].text
like = song.select('span.cnt')[0].text.split()[1]
data = [title, singer, album, like]
data
```

```python
results = []
rank = 0
for song in song_list:
    rank = rank+1
    title = song.select('div.ellipsis.rank01')[0].text.replace('\n', '')
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    album = song.select('div.ellipsis.rank03 > a')[0].text
    like = song.select('span.cnt')[0].text.split()[1]
    data = [rank, title, singer, album, like]
    results.append(data)

results = pd.DataFrame(results)
results
```

```python
results.columns = ['순위','제목','가수','앨범','좋아요']
results
```

```python
results.to_excel('실습_melonchart_top_100.xlsx', index = False)
```
