```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

browser = webdriver.Chrome("c:/r-selenium/chromedriver.exe")
```

```python
results = []
for i in range(1,11) :
    url = f'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={i}'
    browser.get(url)
    html = browser.page_source
    soup = BeautifulSoup(html, 'html.parser')
    channel_list = soup.select('form > table > tbody > tr')
    for channel in channel_list:
        category = channel.select('p.category')[0].text.strip()
        title = channel.select('h1>a')[0].text.strip()
        subscriber = channel.select('td.subscriber_cnt')[0].text
        view = channel.select('td.view_cnt')[0].text
        video = channel.select('td.video_cnt')[0].text
        print(title, category, subscriber, view, video, sep = ' / ')
        data = [title, category, subscriber, view, video]
        results.append(data)
results = pd.DataFrame(results)
results.to_excel('실습_youtube크롤링.xlsx', index = False, header = False)
```
