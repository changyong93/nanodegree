```python
#browser
browser = webdriver.Chrome('C:/r-selenium/chromedriver.exe')
url = 'https://daum.net'
browser.get(url)
```

```python
#url 한글 => 자음 및 모음을 %DD %AA같은 형태로 변환
#url 주소를 보니 %q 뒤에 검색어가 변형되어 들어가는 것으로 보이므로, 검색어 명령해보기
url = 'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q=부동산'
browser.get(url)

#URL = 주소 + 파라미터, 주소 ex)naver.com, daum.net 등 일반적으로 알고 있는 주소
#물음표가 구분자이고 그 뒤가 파라미터고, w=tot, DA=YZR 등이 각 파라미터
#파라미터 속성명 = 값 % 속성명 = 값 &.....
```

```python
# 여러 검색어 찾기
search_list = ['부동산','파이썬','삼성','입시']
for word in search_list:
    url = f'https://search.daum.net/search?w=tot&DA=YZR&t__nil_searchbox=btn&sug=&sugo=&sq=&o=&q={word}'
    
    browser.get(url)
    browser.implicitly_wait(3)

#이후 크롤링 및 파일 저장
```

```python
#웹페이지 정보 다운받기
html = browser.page_source
html
```
